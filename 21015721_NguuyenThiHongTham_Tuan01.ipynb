```python

img = cv.imread('/content/drive/MyDrive/ColabNotebooks/21015721_NguyenThiHongTham_IMG_PROCESSING/anhchandung.png')
# chuyn i ta  nh

def matrix2D(img, angle):
  rota = np.radians(angle)
  width = img.shape[1]
  height = img.shape[0]

  matrix2D = np.zeros((height, width, 3), dtype=np.uint8)
  centerX = width//2
  centerY = height//2
  for i in range (width):
      for j in range (height):
          x = int((i - centerX) * np.cos(rota) - (j - centerY) * np.sin(rota) + centerX)
          y = int((i - centerX) * np.sin(rota) + (j - centerY) * np.cos(rota) + centerY)


          if 0 <= x < width and 0 <= y < height:
            matrix2D[j,i] = img[y,x]

  return matrix2D


r = matrix2D(img, 50)
plt.imshow(r[:,:,::-1], cmap='gray')
plt.axis('off')
plt.show()
```

```python
b,g,r = cv.split(img)
imgMerged = cv.merge((b,g,r))
plt.figure(figsize=(20,15))
plt.subplot(141);plt.imshow(b,cmap='gray');plt.title("Blue channel")
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("Green channel")
plt.subplot(143);plt.imshow(r,cmap='gray');plt.title("Red channel")
plt.subplot(144);plt.imshow(imgMerged[:,:,::-1]);plt.title("Merged Output")
```

```python
import cv2
#  vit bng hm
def img_threshold(img, thres_value):
  img[img >= thres_value] = 255
  img[img < thres_value] = 0
  return img

img = cv.imread('/content/drive/MyDrive/ColabNotebooks/21015721_NguyenThiHongTham_IMG_PROCESSING/IMG/anhdown.jpg')

binary_img = img_threshold(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), 127)
fig, ax = plt.subplots(1, 2, figsize=(10,10))
ax[0].imshow(img[:,:,::-1])
ax[0].set_title('nh gc')
ax[0].axis('off')

ax[1].imshow(binary_img,cmap='gray')
ax[1].set_title('nh trng en')
ax[1].axis('off')
```
