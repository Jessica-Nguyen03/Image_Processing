```python
# 2a

#  tính cường độ gradien sử dụng bộ lọc sobel
matran_kernel_Y = np.array([[ -1, 0, 1],
                            [ -2, 0, 2],
                            [ -1, 0, 1]])
matran_kernel_X = np.array([[ -1, -2, -1],
                            [  0,  0,  0],
                            [  1,  2,  1]])

# nhân tích chập 2 ma trận kernel

x = cv2.filter2D(anh_xam, cv2.CV_64F, matran_kernel_X)
y = cv2.filter2D(anh_xam, cv2.CV_64F, matran_kernel_Y)

# tính cường độ
anh_cuong_do = np.sqrt(np.square(x) + np.square(y))

# # chuẩn hóa kết quả
anh_cuong_do_chuanhoa = anh_cuong_do / np.max(anh_cuong_do) * 255
anh_cuong_do_chuanhoa = anh_cuong_do_chuanhoa.astype(np.uint8)

# hien thi
fx,ax = plt.subplots(1,4,figsize=(15,5))
ax[0].imshow(anh_xam,cmap = 'gray')
ax[0].set_title('Ảnh gốc')
ax[0].axis('off')
ax[1].imshow(x,cmap = 'gray')
ax[1].set_title('Sobel X')
ax[1].axis('off')
ax[2].imshow(y,cmap = 'gray')
ax[2].set_title('Sobel Y')
ax[2].axis('off')
ax[3].imshow(anh_cuong_do_chuanhoa,cmap = 'gray')
ax[3].set_title('Ảnh sau khi tính cường độ gradient với Sobel')
ax[3].axis('off')
plt.tight_layout()
plt.show()
```

```python
# 2b

netanh = cv2.addWeighted(anh_xam, 1.5, anh_cuong_do_chuanhoa, -0.5, 0)
```

```python
# hien thi

fx,ax = plt.subplots(1,3,figsize=(15,5))
ax[0].imshow(anh_xam,cmap='gray')
ax[0].set_title('Ảnh gốc')
ax[0].axis('off')
ax[1].imshow(anh_cuong_do_chuanhoa,cmap = 'gray')
ax[1].set_title('Ảnh sau khi tính cường độ gradient với Sobel')
ax[1].axis('off')
ax[2].imshow(netanh,cmap = 'gray')
ax[2].set_title('Ảnh sau khi làm sắc nét với Sobel')
ax[2].axis('off')
plt.tight_layout()
plt.show()
```

```python

# 2c Viết hàm xác định ngưỡng bằng phương pháp  ngưỡng toàn cục cơ bản
def nguong_toan_cuc(anh):
  anh_coppy = anh.copy()
  t = np.mean(anh)
  g1=[]
  g2 = []
  while(True):
    for i in range (anh_coppy.shape[0]):
      for j in range (anh_coppy.shape[1]):
        if (anh_coppy[i,j] < t):
          g1.append(anh_coppy[i,j])

        else:
          g2.append(anh_coppy[i,j])
    m1 = np.mean(g1)
    m2 = np.mean(g2)

    t0 = (m1+m2)/2
    delta_t = abs(t - t0)
    if (delta_t < 1):
      break
    t = t0
    g1.clear()
    g2.clear()
  return t0
```

```python
# 2d Viết hàm  phân đoạn ảnh  xám  với ngưỡng T xác định được  ở câu c.  Hiển thị ảnh sau phân đoạn
def phan_doan_bang_cat_nguong(anh,nguong): # Định nghĩa hàm phân đoạn bằng cắt ngưỡng
    m, n= anh.shape
    img_phan_doan_cat_nguong = np.zeros([m, n])
    for i in range(m):
      for j in range(n):
          if (anh[i,j] < nguong):
                img_phan_doan_cat_nguong [i,j] = 0
          else:
                img_phan_doan_cat_nguong[i,j] = 225 # tương đương gt 1 trong công thức 1
    return img_phan_doan_cat_nguong
```

```python
anh = cv2.imread('/content/drive/MyDrive/ColabNotebooks/anh.jpg')
anh = cv2.cvtColor(anh,cv2.COLOR_BGR2GRAY)

anh_sau_phan_doan_cat_nguong = phan_doan_bang_cat_nguong(anh,nguong_toan_cuc(anh))
fx,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(anh,cmap = 'gray')
ax[0].set_title('Ảnh gốc')
ax[0].axis('off')
ax[1].imshow(anh_sau_phan_doan_cat_nguong,cmap = 'gray')
ax[1].set_title('Ảnh sau khi phân đoạn bằng cắt ngưỡng')
ax[1].axis('off')
plt.tight_layout()
plt.show()
```
