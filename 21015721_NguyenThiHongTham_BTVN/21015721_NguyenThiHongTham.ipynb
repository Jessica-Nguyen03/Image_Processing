```python
read_img = cv.imread('/content/gdrive/MyDrive/IMG_PROCESSING/IMG/21015721_NguyenThiHongTham_BT/anhchandung.png',0)
```

```python
img_gray_4bit = (read_img//16)
img_gray_2bit = (read_img//64)
img_gray_1bit = (read_img//128)

plt.figure(figsize=(15,15))

plt.subplot(1,4,1)
plt.title('Ảnh xám gốc')
plt.imshow(read_img, cmap='gray')

plt.subplot(1,4,2)
plt.title('Ảnh xám 4 bit')
plt.imshow(img_gray_4bit, cmap='gray')


plt.subplot(1,4,3)
plt.title('Ảnh xám 2 bit')
plt.imshow(img_gray_2bit, cmap='gray')


plt.subplot(1,4,4)
plt.title('Ảnh xám 1 bit')
plt.imshow(img_gray_1bit, cmap='gray')
plt.show()
```
