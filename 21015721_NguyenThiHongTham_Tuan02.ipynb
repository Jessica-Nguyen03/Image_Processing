#### Read an image and resize it to 512x512 using the appropriate function. Then down sample the image by 4 so that the final size of the image is 128x128. Display and save the image to the disk.

```python
img = cv.imread('/content/drive/MyDrive/ColabNotebooks/21015721_NguyenThiHongTham_Lab01/ironman.jpg')
#resize img

def resize_img(img,newsize):

  old_size = img.shape[:2]

  scale_x = float(newsize[0]) / old_size[0]
  scale_y = float(newsize[1]) / old_size[1]

  new_img = np.zeros((newsize[0],newsize[1],img.shape[2]), dtype=img.dtype)
  for i in range(new_img.shape[0]):
    for j in range(new_img.shape[1]):
      x = int(i/scale_x)
      y = int(j/scale_y)
      new_img[i,j] = img[x,y]


  return new_img
img = resize_img(img, (512,512))

img_new = resize_img(img, (128,128))

fx,ax = plt.subplots(1,2,figsize=(15,15))
ax[0].imshow(img)
ax[0].title.set_text('512x512')
ax[0].axis('off')
ax[1].imshow(img_new)
ax[1].title.set_text('128x128')
ax[1].axis('off')

plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

def create_image_with_colored_corners(size=500):
    # Create a white image of the specified size
    image = np.ones((size, size, 3), dtype=np.uint8) * 255

    # Calculate the size of the colored boxes (1/8th of the image size)
    box_size = size // 8

    # Top-left corner: Red box
    image[:box_size, :box_size] = [255, 0, 0]  # Red color (R, G, B)

    # Top-right corner: Green box
    image[:box_size, -box_size:] = [0, 255, 0]  # Green color (R, G, B)

    # Bottom-left corner: Blue box
    image[-box_size:, :box_size] = [0, 0, 255]  # Blue color (R, G, B)

    # Bottom-right corner: Black box
    image[-box_size:, -box_size:] = [0, 0, 0]  # Black color (R, G, B)

    # Display the image
    plt.imshow(image)
    plt.axis('off')
    plt.show()

# Example usage
create_image_with_colored_corners(500)
```

```python
def draw_color():
  matrix = np.ones([5,5,3], dtype=np.uint8)
  width = 5
  height = 5
  for i in range (matrix.shape[0]):
    for j in range (matrix.shape[1]):
        if i == 0 and j == 0 or i == width - 1 and j == height -1 or i == width - 1 and j == 0 or i == 0 and j == height -1:

                  if i == 0 and j == 0:
                    matrix[i,j] = [255,0,0]
                  if i == width - 1 and j == height -1:
                    matrix[i,j] = [0,0,0]

                  if i == width - 1 and j == 0:
                    matrix[i,j] = [0,0,255]


                  if i == 0 and j == height -1:
                    matrix[i,j] = [0,255,0]

        else:
          matrix[i,j] = [255,255,255]
  return matrix

grayscale = draw_color()

plt.imshow(grayscale)
plt.xlabel('x')
plt.ylabel('y')

plt.axis('off')
plt.show()
```
